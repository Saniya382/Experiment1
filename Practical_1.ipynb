```python
# Program to demonstrate all major List operations in Python
```

```python
# 1. Creating a List
# -----------------------------
my_list = [10, 20, 30, 40, 50]
print("Initial List:", my_list)
```

```python
# 2. Accessing Elements
# -----------------------------
print("First element:", my_list[0])
print("Last element:", my_list[-1])
```

```python
# 3. Slicing a List
# -----------------------------
print("Elements from index 1 to 3:", my_list[1:4])
print("First three elements:", my_list[:3])
print("Elements from index 2 to end:", my_list[2:])
```

```python
# 4. Adding Elements
# -----------------------------
# append() - adds element at the end
my_list.append(60)
print("After append:", my_list)

# insert() - adds element at specific position
my_list.insert(2, 25)
print("After insert at index 2:", my_list)

# extend() - adds multiple elements
my_list.extend([70, 80, 90])
print("After extend:", my_list)
```

```python
print(my_list)
```

```python
# 5. Removing Elements
# -----------------------------
# remove() - removes specific element
my_list.remove(30)
print("After remove 30:", my_list)

# pop() - removes element at given index
my_list.pop(3)
print("After pop at index 3:", my_list)

# clear() - removes all elements
temp_list = my_list.copy()
temp_list.clear()
print("After clear:", temp_list)
```

```python
# 6. Updating Elements
# -----------------------------
my_list[1] = 300
print("After updating index 1:", my_list)
```

```python
# 7. Searching in List
# -----------------------------
print("Index of element 80:", my_list.index(80))
print("Count of element 50:", my_list.count(50))
```

```python
# 8. Sorting the List
# -----------------------------
my_list.sort()
print("After sorting:", my_list)

# reverse sorting
my_list.sort(reverse=True)
print("After reverse sorting:", my_list)
```

```python
# 9. Reversing the List
# -----------------------------
my_list.reverse()
print("After reversing:", my_list)
```

```python
# 10. Copying a List
# -----------------------------
new_list = my_list.copy()
print("Copied list:", new_list)
```

```python
# 11. Length of List
# -----------------------------
print("Length of list:", len(my_list))
```

```python
# 12. Looping through List
# -----------------------------
print("Elements using for loop:")
for item in my_list:
    print(item)
```

```python
# 13. Checking Membership
# -----------------------------
print("Is 40 in list?", 40 in my_list)
print("Is 100 in list?", 100 not in my_list)
```

```python
# 14. List Concatenation
# -----------------------------
list1 = [1, 2, 3]
list2 = [4, 5, 6]
combined_list = list1 + list2
print("Concatenated list:", combined_list)
```

```python
# 15. Deleting List
# -----------------------------
del list1
print("List operations completed successfully.")
```
